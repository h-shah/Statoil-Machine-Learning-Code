```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
import matplotlib.pyplot as plt
import pickle as pkl
%matplotlib inline
```

Load in data and check types and sizes

```python
train = pd.read_json('D:/Libraries/Documents/Tensorflow/tensorflow/Statoil/train.json')
train['inc_angle'] = pd.to_numeric(train['inc_angle'],errors='coerce')
test = pd.read_json('D:/Libraries/Documents/Tensorflow/tensorflow/Statoil/test.json')
test['inc_angle'] = pd.to_numeric(test['inc_angle'],errors='coerce')
```

```python
type(train)
train.head()
```

Converting to numpy arrays

```python
band1 = np.ndarray(shape=[1604,5625],dtype=float)
band2 = np.ndarray(shape=[1604,5625],dtype=float)
```

```python
for i in range(1604):
    band1[i] = np.array(train["band_1"][i])
    band2[i] = np.array(train['band_2'][i])
```

```python
band1.dtype
```

```python
print(band1[0:3,0:3])
print(band2[0:3,0:3])
```

```python
angles = np.array(train['inc_angle'])
print(np.sum(np.isnan(angles)))
```

```python
labels = np.array(train['is_iceberg'])
print(np.sum(labels))
```

```python
ids = np.array(train['id'])
ids.shape
```

Let's do some visualization

```python
fig = plt.figure(1,figsize=(15,15))
icebergsb1 = band1[labels==1]
for i in range(9):
    ax = fig.add_subplot(3,3,i+1)
    arr = np.reshape(np.array(icebergsb1[i]),(75,75))
    ax.imshow(arr,cmap='inferno')
    
plt.show()
```

```python
fig = plt.figure(1,figsize=(15,15))
shipsb1 = band1[labels==0]
for i in range(9):
    ax = fig.add_subplot(3,3,i+1)
    arr = np.reshape(np.array(shipsb1[i]),(75,75))
    ax.imshow(arr,cmap='inferno')
    
plt.show()
```

```python
shuffle = np.random.permutation(range(1604))
```

```python
b1train = band1[shuffle,:][:1404,:]
b1valid = band1[shuffle,:][1404:1504,:]
b1test = band1[shuffle,:][1504:,:]

b2train = band2[shuffle,:][:1404,:]
b2valid = band2[shuffle,:][1404:1504,:]
b2test = band2[shuffle,:][1504:,:]

anglestrain = angles[shuffle][:1404]
anglesvalid = angles[shuffle][1404:1504]
anglestest = angles[shuffle][1504:]

labelstrain = labels[shuffle][:1404]
labelsvalid = labels[shuffle][1404:1504]
labelstest = labels[shuffle][1504:]

idstrain = ids[shuffle][:1404]
idsvalid = ids[shuffle][1404:1504]
idstest = ids[shuffle][1504:]
```

Pickle the curated data

```python
data = {}
data['b1train']=b1train
data['b1valid']=b1valid
data['b1test']=b1test
data['b2train']=b2train
data['b2valid']=b2valid
data['b2test']=b2test
data['anglestrain']=anglestrain
data['anglesvalid']=anglesvalid
data['anglestest']=anglestest
data['labelstrain']=labelstrain
data['labelsvalid']=labelsvalid
data['labelstest']=labelstest
data['idstrain']=idstrain
data['idsvalid']=idsvalid
data['idstest']=idstest

with open('D:/Libraries/Documents/Tensorflow/tensorflow/Statoil/labeled_data.pkl', 'wb') as pklfile:
    pkl.dump(data, pklfile)
```

```python
b1unlab = np.ndarray(shape=[rows,5625],dtype=float)
b2unlab = np.ndarray(shape=[rows,5625],dtype=float)

for i in range(rows):
    b1unlab[i] = np.array(test["band_1"][i])
    b2unlab[i] = np.array(test["band_2"][i])

print("band_1")
print(b1unlab[0:3,0:3])
print("band_2")
print(b2unlab[0:3,0:3])

anglesunlab = np.array(test['inc_angle'])
print('NaN angles: ' + str(np.sum(np.isnan(anglesunlab))))
print(anglesunlab[0:5])

idsunlab = np.array(test['id'])
print(idsunlab[0:5])
```

```python
unlabeled_data = {}
unlabeled_data['b1unlab']=b1unlab
unlabeled_data['b2unlab']=b2unlab
unlabeled_data['anglesunlab']=anglesunlab
unlabeled_data['idsunlab']=idsunlab

with open('D:/Libraries/Documents/Tensorflow/tensorflow/Statoil/unlabeled_data.pkl', 'wb') as f:
    pkl.dump(unlabeled_data, f)
```
